```python
import pandas as pd
```

```python
url = 'http://bit.ly/kaggletrain'
train = pd.read_csv(url)
train.head(3)
```

```python
#map function being used to map the values
#1:1 relation is established using map()

#CREATE A NEW COLUMN
train['Sex_num'] = train.Sex.map({'female':0, 'male':1})
```

```python
#looking at the map that maps male:1 and female:0
train.loc[0:10, ['Sex', 'Sex_num']]
```

```python
#eg. if we want to calculate the length of string in 'Name' column (using python's length function)

#creating a new column and finding the results
#apply() applies a function to each element in the series
train['Name_length'] = train.Name.apply(len) 
```

```python
train.loc[0:8, ['Name', 'Name_length']]
```

```python
import numpy as np

#say we look at far column and we want to round it up
#using numpy's ceil() to round it up
#create another column that does saves values for you
train['Fare_round_up_value'] = train.Fare.apply(np.ceil)
```

```python
train.loc[0:5, ['Fare', 'Fare_round_up_value']]
```

```python
#if you wish to extract the last name of each person:
#lets see what each cell in the 'Name' column has
train.head(10)
```

```python
#using a str method
#series is a list of string
#each cell has two strings in a list
train.Name.str.split(',').head()
```

```python
#we just want the 1st string from the list
#we create a function to retrieve
def get_element(my_list, position):
    return my_list[position]
```

```python
#pass position = 1
train.Name.str.split(',').apply(get_element, position = 0).head()
```

```python
#instead, we can use a lambda function
#input x - the list in this case
#output[0] - the first string in the list in this case
train.Name.str.split(',').apply(lambda x: x[0]).head()
```

```python
#if you want to get the second string
train.Name.str.split(',').apply(lambda x: x[1]).head()
```

```python
#apply() function as a dataframe method - applies function on either axis of the frame
url = 'http://bit.ly/drinksbycountry'
drinks = pd.read_csv(url)
drinks.head()
```

```python
drinks.loc[:, 'beer_servings':'wine_servings'].head()
```

```python
drinks.loc[:, 'beer_servings':'total_litres_of_pure_alcohol'].head()
```

```python
#you want to apply() method to travel axis=0 (downwards in a column)
#apply python's max() function
drinks.loc[:, 'beer_servings':'wine_servings'].apply(max, axis=0)
```

```python
#if you want to apply() methos to axis=1 (right in a row)
#applying python's max() function again
drinks.loc[:, 'beer_servings':'wine_servings'].apply(max,axis=1)
```

```python
#finding which column is the maximum's category name
#What is argmax()? - : you have a set of candidates C, 
#a score function f, and you want to compute 
#an optimal candidate c in C according to f, i.e. one that maximizes f(c).
drinks.loc[:,'beer_servings':'wine_servings'].apply(np.argmax, axis=1)
```

```python
#applymap() function to every element 
drinks.loc[:, 'beer_servings': 'wine_servings'].applymap(float).head()
```

```python
#lets overwrite existing table
drinks.loc[:, 'beer_servings': 'wine_servings'] = drinks.loc[:, 'beer_servings': 'wine_servings'].applymap(float)
drinks.head()
```
